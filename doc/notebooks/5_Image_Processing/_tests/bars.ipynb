```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from scipy import ndimage
import pandas as pd
import os
%matplotlib nbagg
```

```python
path = "balls.jpg"
files = os.listdir("./")
if path in files:
    print("Ok, the file is in {0}".format(files))
else:
    print("The file is not in {0} , retry !".format(files))
```

```python
im = Image.open(path)
Nc, Nl = im.size
im = im.resize((Nc // 4 ,Nl // 4),Image.ANTIALIAS)
fig, ax = plt.subplots()
#ax.axis("off")
plt.imshow(im)
plt.colorbar()
plt.show()
```

```python
R, G, B = im.split()
R = np.array(R)
G = np.array(G)
B = np.array(B)
```

## 1) Histogram 

Plot the histogram of the image and find the right level for the thresholding.


```python
plt.figure()
plt.hist(R.flatten(), bins = np.arange(256), histtype = "step", color = "r")
plt.hist(G.flatten(), bins = np.arange(256), histtype = "step", color = "g")
plt.hist(B.flatten(), bins = np.arange(256), histtype = "step", color = "b")
plt.grid()
plt.show()
```

## 2) Thresholding

Use thresholding to convert the image to binary format.

```python
Bt = B < 50
plt.figure()
plt.imshow(Bt, cmap = cm.gray)
plt.colorbar()
plt.show()
```

## 3) Labeling

Use the `ndimage.measurements.label` to label each grain and plot the result.

```python
Btc = ndimage.binary_closing(Bt, structure = np.ones((3, 3)))


Bl, number  = ndimage.measurements.label(Btc)
plt.figure()
plt.imshow(np.where(Bl !=0, Bl, np.nan), cmap = cm.jet)
plt.show()
number
```

```python
obj = ndimage.measurements.find_objects(Bl)
len(obj)
```

```python
data = pd.DataFrame(columns = ["A", "xg", "yg", "Ixx", "Iyy", "Ixy", "dx", "dy", "theta"])
for i in range(len(obj)):
    y, x = np.where(Bl == i+1)
    xg, yg = x.mean(), y.mean()
    x = x - xg
    y = y - yg
    A = len(x)
    Ixx = (y**2).sum()
    Iyy = (x**2).sum()
    Ixy = (x*y).sum()
    I = np.array([[Ixx, -Ixy], [-Ixy, Iyy]])
    eigvals, eigvecs = np.linalg.eig(I)
    loc = np.argsort(eigvals)[::-1]
    d = eigvecs[loc[0]]
    theta =  np.degrees(np.arccos(d[1]))
    
    data.loc[i+1] = [A, xg, yg, Ixx, Iyy, Ixy, d[0], d[1], theta]
data.sort_values("A", inplace = True, ascending = False)

data
```

```python
fig = plt.figure()
counter = 1
for i in data.index.values:
    ax = fig.add_subplot(3,4, counter)
    z = Image.fromarray(np.array(im)[obj[i-1]])
    z = z.rotate(data.loc[i, "theta"], expand = True)
    z = np.array(z)
    plt.imshow(z)
    plt.title(str(i))
    ax.axis("off")
    counter += 1
plt.show()    
```
