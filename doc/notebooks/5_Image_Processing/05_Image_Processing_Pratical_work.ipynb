# Practical work on Hubble Deep Field
## Setup

```python
%matplotlib nbagg
from skimage import data
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import cm
import numpy as np
```

```python
image = data.hubble_deep_field()[0:500, 0:500]
channels = ['red', 'green', 'blue']

fig = plt.figure("HDF")
plt.imshow(image)
plt.axis("OFF")
plt.show()
```

```python
fig = plt.figure("Channels")
for i in range(0,3):
    fig.add_subplot(1, 3, i+1)
    plt.imshow(image[:,:,i], origin = "upper", cmap = cm.gray)
    plt.title("{0}".format((channels[i])))
    plt.axis("OFF")
plt.show()
```

```python
fig = plt.figure("Histograms")
for i in range(0,3):
    fig.add_subplot(1, 3, i+1)
    plt.hist(image[:,:,i].flatten(), bins=np.arange(256), histtype = "stepfilled", color="{0}".format(channels[i]))
    plt.title("{0}".format(channels[i]))
    plt.grid()
    plt.xlabel("Pixel value")
    plt.ylabel("Pixel count")
plt.show()
```

```python
fig = plt.figure("Thresholding")

for i in range(0,3):
    fig.add_subplot(1, 3, i+1)
    plt.imshow(np.where(image[:,:,i]<100,np.nan,1), origin = "upper")
    plt.title("{0}".format(channels[i]))
plt.show()
```
