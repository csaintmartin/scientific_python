## Introduction with a simple tow points line equation

Let's annalyse how to determine the simple equation of a linear function ($y=ax+b$) kowing tow points of the line.

```python
# the known points
y=np.array([-1,2])
x=np.array([.5,3])
x,y
```

```python
fig = plt.figure()
plt.plot(x,y,'o')
plt.grid()
```

Todertermine the $a$ and $b$ coefficents, we have to solve the following system :

$y_0=ax_0+b$

$y_1=ax_1+b$


Lets reorgenized a bit this system :

$ b + ax_0 = y_0 $

$ b + ax_1 = y_1 $

Its konw easy to write it as a linear systeme :

$
\begin{bmatrix} 
1 & x_0 \\
1 & x_1 
\end{bmatrix}
.
\begin{bmatrix} 
b \\
a
\end{bmatrix}
=
\begin{bmatrix} 
y_0 \\
y_1
\end{bmatrix}
$

Therfore the $a$ and $b$ coefficents can be determine by inversing the linear system:

$
\begin{bmatrix} 
b \\
a
\end{bmatrix}
=
\begin{bmatrix} 
1 & x_0 \\
1 & x_1 
\end{bmatrix}^{-1}
.
\begin{bmatrix} 
y_0 \\
y_1
\end{bmatrix}
$

### Let's do this with python

```python
# definition of the involved matrix :
X = np.array([[1, x[0]],[1,x[1]]])
Y = np.array([y[0],y[1]]).T
```

```python
# Inversion of the systeme
coef=np.dot(np.linalg.inv(X),Y)
a = coef[1]
b = coef[0]
print("a=",a,)
print("b=",b)
```

```python
# draw the obtained line
N = 10
xmin, xmax = 0., 4
xi = np.linspace(xmin, xmax, N)
yi = a * xi + b
```

```python
fig = plt.figure()
plt.plot(x,y,'o')
plt.plot(xi,yi,'r')
plt.grid()
```

## What append if we add a third point ?

```python
# the known points
y=np.array([-1,2,0])
x=np.array([.5,3,1.5])
x,y
```

Todertermine the $a$ and $b$ coefficents, we have to solve the following system :

$y_0=ax_0+b$

$y_1=ax_1+b$

$y_2=ax_2+b$

This system is over deffined: we have 2 unkowns ($a$,$b$) and 3 equations. The goal is now to find $a$ and $b$ coefficient (i.e. to find the linear equation) that fit the best the given data.

We can rewrite the equation by to the following linear system:

$
\begin{bmatrix} 
1 & x_0 \\
1 & x_1 \\
1 & x_2
\end{bmatrix}
\begin{bmatrix} 
b \\
a
\end{bmatrix}
=
\begin{bmatrix} 
y_0 \\
y_1 \\
y_2
\end{bmatrix}
$

The system can be express in a genral form :

$X \beta=Y$

The matrix 
$X$ is not square, and therfore not inversible. The methode presented above is no more apllicable.

The trick is to multiply the system by the transposed of $X$ ($X^t$).

$X^t X \beta= X^t Y$

The therm $X^t X$ is now a square matrix. 



Let's check this affirmation:

```python
# Construcion of the X matrix
X = np.array([np.ones(x.size), x]).T
X
```

```python
# Computation of Xt X
np.dot(X.T,X)
```

The obtained matrix is a squart matrix and it can proved that this matrix is symmetric, and therfore inversible.

The $\beta$ vector can be express by :

$\beta= \left( X^t X \right)^{-1} X^t Y$

This operator is the least square estimator.

### Let's do this with python

```python
# Construcion of the X matrix
X = np.array([np.ones(x.size), x]).T

# Construcion of the Y matrix
Y = y.T

print("X=",X)

print("Y=",Y)
```

```python
# Computation of the least square estimator
beta = np.dot(np.linalg.inv(np.dot(X.T,X)),np.dot(X.T,Y))
print("beta=",beta)
```

### Now its time to plot the line

```python
# draw the obtained line
N = 10
xmin, xmax = 0., 4
xi = np.linspace(xmin, xmax, N)
yi = beta[0] + beta[1] * xi
```

```python
fig = plt.figure()
plt.plot(x,y,'o')
plt.plot(xi,yi,'r')
plt.grid()
```

## With a larger data set ?

### The data set : a synthetique data set with some noise

```python
N = 100 
xmin, xmax = 0., 4
x = np.linspace(xmin, xmax, N) + np.random.rand(N)
y = (3.3 * x -2) + 2*np.random.randn(N)
```

```python
fig = plt.figure()
plt.plot(x,y,'o',label = "Data")
plt.legend()
plt.grid()
```

### Determination of the coefficent by regression

```python
# Construcion of the X matrix
X = np.array([np.ones(x.size), x]).T

# Construcion of the Y matrix
Y = y.T

# Computation of the least square estimator
beta = np.dot(np.linalg.inv(np.dot(X.T,X)),np.dot(X.T,Y))

print("The fitted linear function is:")
print("y=",beta[1],"x +",beta[0])
```

### Draw the obtaine regression line

```python
N = 10
xmin, xmax = 0., 5
xi = np.linspace(xmin, xmax, N)
yi = beta[0] + beta[1] * xi
```

```python
fig = plt.figure()
plt.plot(x,y,'o',label = "Data")
plt.plot(xi,yi,'r',label = "Regression")
plt.legend()
plt.grid()
```

## Does python  know how to do that in one line ?

### Of course yes ! 
 let's use the polyfit function of numpy.

```python
# compute the coefficient using the least square methode
coef=np.polyfit(x, y, 1)

# creat the associate 1d polynomila function
fit = np.poly1d(coef)
```

### Draw the obtaine regression line

```python
N = 10
xmin, xmax = 0., 5
xi = np.linspace(xmin, xmax, N)
yi_numpy = fit(xi)
```

```python
fig = plt.figure()
plt.plot(x,y,'o',label = "Data")
plt.plot(xi,yi,'r',label = "Regression")
plt.plot(xi,yi_numpy,'g',label = "Regression Numpy")
plt.legend()
plt.grid()
```

Both regression are equal.
