### Niveaux de gris

```python
gris = np.random.random((5, 15))*256.
fig1 = plt.figure("50 shades of grays")
plt.clf()
ax0 =  fig1.add_subplot(2,1,1)

plt.imshow(gris, origin = "upper", cmap = cm.gray)
plt.ylabel("pixels")
cbar = plt.colorbar()
cbar.set_label("Niveaux de gris")

oeil=np.matrix([[0, 100, 100, 100, 100, 100], 
                [255, 101, 102, 103, 104, 105]])

ax1 =  fig1.add_subplot(2,1,2)
plt.imshow(oeil, origin = "upper", cmap = cm.gray)
plt.xlabel("pixels")
plt.ylabel("pixels")
cbar = plt.colorbar()
cbar.set_label("Niveaux de gris")
plt.show()
```
