**Question 5**: What is the ratio between accidents taking place during day, night, and tranision day/night ? Show results in a graphical way.

```python
dlum = df["lum"].value_counts(normalize=True)
dlum = dlum.sort_index()
```

```python
dlum[3] = dlum[3:5].sum()
dlum
```

```python
val = dlum[1:3].values
explode = (0.1, 0.1, .1)  # only "explode" the 2nd slice (i.e. 'Hogs')
labels = "day", "transition", "night"
fig1, ax1 = plt.subplots()
ax1.pie(val, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

**Question 6**: Propose a graphical repesentation to higlight the geographical position on each accident which have take place on the metropolitan territory.

```python
# Step by step
dfp = df[df.gps=='M']
dfp = dfp[['lat','long']]
dfp = dfp[(dfp.long!=0.0) & (dfp.lat!=0.0)]
dfp.head()
```

```python
plt.figure()
dfp.plot.scatter(x='long', y='lat',s=1);
```

```python
# in one step
plt.figure()
df[(df.long!=0.0) & (df.lat!=0.0) & (df.gps=='M')].plot.scatter(x='long', y='lat',s=.5)
```
