```python
#setup
%load_ext autoreload
%matplotlib nbagg
%autoreload 2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
```

## Introduction ...

### The data set : a synthetique data set

```python
N = 100
xmin, xmax = 0., 4
x = np.linspace(xmin, xmax, N)
y = (x**2-3*x+2*np.sin(5*x))
```

```python
fig = plt.figure()
plt.plot(x,y,'o',label = "numerical data")
plt.legend()
plt.grid()
```
